```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Score': [55, 72, 67, 88, 90, 79, 68, 77, 85, 60]
}
df = pd.DataFrame(data)

# Z-score normalization
def z_score_normalization(column):
    mean = column.mean()
    std = column.std()
    return (column - mean) / std

# Apply normalization
df['Z-Score Normalized'] = z_score_normalization(df['Score'])

# Display results
print("Original and Z-score Normalized Data:")
print(df)

# Visualization
plt.figure(figsize=(8, 4))
plt.plot(df['Score'], label='Original')
plt.plot(df['Z-Score Normalized'], label='Z-score Normalized')
plt.title('Z-Score Normalization')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Topic 6 â€“ Z-Score Normalization

Z-score normalization (also known as standardization) transforms data to have a mean of 0 and a standard deviation of 1. This is especially useful when comparing features with different units or scales.

**Formula:**  
\[
z = \frac{x - \mu}{\sigma}
\]

Where:  
- \( x \) = original value  
- \( \mu \) = mean of the column  
- \( \sigma \) = standard deviation
