```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate 3D synthetic data
np.random.seed(42)
X = np.random.rand(100, 3) * 100

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame for visualization
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(df_pca["PC1"], df_pca["PC2"], c='green', edgecolor='k')
plt.title("PCA – Dimensionality Reduction (3D to 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
### Topic 8 – Principal Component Analysis (PCA)

In this notebook, we apply **Principal Component Analysis (PCA)** to reduce the dimensionality of a 3D dataset to 2D.

Steps performed:
1. **Standardization**: We scaled the data using `StandardScaler` to ensure that PCA is not biased by magnitude.
2. **PCA transformation**: We reduced the original 3D features to 2 principal components using `sklearn.decomposition.PCA`.
3. **Visualization**: The result is plotted on a 2D scatter plot for better interpretation.

PCA helps with:
- Dimensionality reduction,
- Noise reduction,
- Visualization of high-dimensional data.

This technique is especially useful before clustering or classification tasks.
```
