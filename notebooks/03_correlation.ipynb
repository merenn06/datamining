```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed
np.random.seed(42)

# Correlated data
x = np.arange(100)
y = x + np.random.normal(0, 10, 100)  # y is highly correlated with x

# Uncorrelated data
z = np.random.normal(0, 50, 100)  # Random noise, not correlated with x

# Create DataFrame
df = pd.DataFrame({
    "x": x,
    "y (correlated)": y,
    "z (uncorrelated)": z
})

# Compute correlation coefficients
corr_y = np.corrcoef(x, y)[0, 1]
corr_z = np.corrcoef(x, z)[0, 1]

# Print correlation values
print(f"Correlation between x and y: {corr_y:.2f}")
print(f"Correlation between x and z: {corr_z:.2f}")

# Plot
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=x, y=y)
plt.title(f"Correlated Data (r = {corr_y:.2f})")

plt.subplot(1, 2, 2)
sns.scatterplot(x=x, y=z)
plt.title(f"Uncorrelated Data (r = {corr_z:.2f})")

plt.tight_layout()
plt.show()
```

### Topic 3 â€“ Correlated and Uncorrelated Data

In this notebook, we analyze and visualize two types of relationships between variables:

- **Correlated data**: two variables that show a clear linear relationship.
- **Uncorrelated data**: variables that do not follow any pattern or association.

We used:
- `numpy.corrcoef()` to compute Pearson's correlation coefficient (r).
- `seaborn.scatterplot()` to plot the data visually.

**Observations:**
- The correlation between `x` and `y` is high (positive), close to +1.
- The correlation between `x` and `z` is near zero, indicating no linear relationship.
