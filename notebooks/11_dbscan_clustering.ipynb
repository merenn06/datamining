```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate sample data (moons shape for better DBSCAN effect)
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

### Topic 11 – DBSCAN Clustering

In this notebook, we use **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** to group data based on density.

Why DBSCAN?
- No need to specify the number of clusters,
- Can detect arbitrarily shaped clusters,
- Can identify and ignore noise (outliers).

Steps:
1. We generated non-linear moon-shaped data using `make_moons`.
2. Standardized the features with `StandardScaler`.
3. Applied `DBSCAN` with `eps=0.3` and `min_samples=5`.

Clusters are colored differently. Noise points are usually labeled as `-1`.
