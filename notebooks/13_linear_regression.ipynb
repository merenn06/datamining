```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create simple linear data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted Line')
plt.title("Linear Regression – Simple Model")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Print coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")
```

### Topic 13 – Linear Regression

In this notebook, we perform **Simple Linear Regression** using `sklearn.linear_model.LinearRegression`.

Steps:
1. Created simple (X, y) data with a roughly linear pattern.
2. Trained a regression model to learn the best-fit line.
3. Visualized both actual data points and the predicted line.

Key outputs:
- **Intercept**: where the line crosses the y-axis,
- **Slope**: how much y changes when x increases by 1.

Linear regression is used in:
- Predicting trends (e.g., sales, stock prices),
- Estimating relationships between variables,
- Feature importance in machine learning.
