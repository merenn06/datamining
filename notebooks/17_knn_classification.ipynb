```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train KNN model (k=3)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict
y_pred = knn.predict(X_test)

# Report
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Confusion Matrix
ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test, display_labels=iris.target_names, cmap='Purples')
plt.title("KNN Confusion Matrix")
plt.grid(False)
plt.show()
```

### Topic 17 – K-Nearest Neighbors (KNN)

In this notebook, we use **K-Nearest Neighbors (KNN)** to classify the Iris dataset.

Steps:
1. Loaded data using `load_iris()` from `sklearn.datasets`.
2. Split data into training and test sets.
3. Trained a KNN classifier with `k=3`.
4. Predicted test set and evaluated results.

KNN works by:
- Finding the `k` closest points (neighbors) to a new sample,
- Assigning the most common class among them.

Evaluation is done using a classification report and confusion matrix.

Use cases:
- Handwriting recognition,
- Recommender systems,
- Anomaly detection.
